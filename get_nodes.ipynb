```python
import requests
from bs4 import BeautifulSoup
#/?column=percent_used_gpu_memory&order=asc
LOCATIONS = ['lab325', '475']
MAX_GPU_USAGE = 5
MAX_CPU_USAGE = 5
NAME = 0
GPU_USAGE =7
LOCATION = 1
CPU_USAGE = 4
```

```python
def parse_row(row):
    new_row = []
    for col in row.findAll('td'):
        #print('col:', col)
        new_row.append(col.text)
    return new_row
```

```python
def get_parsed_table():
    URL = 'https://www.cs.colostate.edu/machinestats'
    PARAMS = {'column': 'percent_used_gpu_memory', 'order':'asc'}
    r = requests.get(url = URL, params=PARAMS)
    soup = BeautifulSoup(r.text, 'html.parser')
    html_table = soup.table.findAll('tr')
    parsed_table = []
    for row in html_table[1:]:
        parsed_table.append(parse_row(row))
    return parsed_table
```

```python
table = get_parsed_table()
```

```python
def get_machines():
    table = get_parsed_table()
    acceptable_machines = []
    for machine in table:
        if machine[GPU_USAGE] != 'N/A' and machine[GPU_USAGE] !='Retired' and float(machine[GPU_USAGE]) < MAX_GPU_USAGE and float(machine[CPU_USAGE]) <MAX_CPU_USAGE:
            acceptable_machines.append(machine[NAME])
    return acceptable_machines
```

```python
def format_machines(machines):
    s = ''
    for m  in machines:
        s += m + " "
    return s
```

```python
acc_mach = get_machines()
acc_mach
```

```python
format_machines(acc_mach)
```
